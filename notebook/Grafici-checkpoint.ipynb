```python
#import e parsing file di log

def lettura_log(path):

    file = open(path, 'r')
    Lines = file.readlines()

    coppie = []

    for line in Lines:
        if 'COUPLE(DISTANCE RMSE)' in line:
            split = line.split(':')
            s = split[4].replace('[','')
            s = s.replace('(','')
            s = s.replace(')','')
            s = s.replace(']','')
            s = s.split(',')
            a = int(s[0])
            b = float(s[1])
            coppie.append((a,b))

    return coppie

```

```python
#funzione per graficare

import matplotlib.pyplot as plt

def grafico(path):

    coppie = lettura_log(path)

    x_val = [x[0] for x in coppie]
    y_val = [x[1] for x in coppie]

    fig, ax = plt.subplots(figsize=(10, 5))

    ax.plot(x_val,y_val)
    ax.plot(x_val,y_val,'or')

    ax.set_ylabel('RMSE Distance (Gurobi -TensorFlow)')
    ax.set_xlabel('Number of Iteration')

    return plt.show()
```

**Classe** : Setosa

**Parametri** : C=1, GaussianKernel(0.25), penalization = 10

**N_iter** 100 -> 1500
